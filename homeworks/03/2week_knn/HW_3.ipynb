##### Реализуйте KNN. Сами, как делали это на лекции. 

Можете улучшить его.
(например использовать kd-tree для быстрого поиска ближайших соседей, 
 или обходить соседей и класть расстояния в мин-кучу - тогда вы быстрее будете набирать их, без сортировки.)

```python
class KNN:
    """
    simple KNN classifier with O(n) complexity for memory,
    and O(n log n) complexity - for nn search.
    """
    
    def __init__(self, n):
        self.n = n
        
    def fit(self, X, y):
        self.X = X
        self.y = y
    
    def predict(self, X):
        y_hat = []
        assert len(X.shape) == 2
        for str_ in X:
            
            dist = 0.
            for i, val in enumerate(str_):
                dist += (self.X[:, i] - val)**2
            dist = np.sqrt(dist)
            
            idx = np.argsort(dist)[:self.n]
            counts = np.bincount(self.y[idx].astype(int))
            prediction = np.argmax(counts)
            y_hat.append(prediction)
            
        return np.array(y_hat)
```

```python
knn = KNN(5)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
y_pred
```

```python
df_titanic = pd.read_csv('titanic.csv')
```

```python
columns = ['Sex', 'Age', 'Pclass', 'Survived', 'Fare', 'Embarked', 'Cabin']
df_titanic = df_titanic[columns]
```

```python
def sex_to_binary(item):
    if item == 'male':
        return 1
    elif item == 'female':
        return 0
    else:
        pass
```

```python
df_titanic['Sex'] = df_titanic['Sex'].apply(sex_to_binary)
df_titanic = df_titanic.fillna({'Age' : df_titanic['Age'].mean(), 'Embarked' : df_titanic['Embarked'].mode()[0]})
df_titanic['Cabin'] = df_titanic['Cabin'].apply(lambda x: x[0] if type(x)==str else np.nan)

for pas_class in range(1,4):
    df_titanic[df_titanic['Pclass'] == pas_class] = df_titanic[df_titanic['Pclass'] == pas_class].fillna(value = df_titanic[df_titanic['Pclass'] == pas_class]['Cabin'].mode()[0])

```

```python
df_titanic['Cabin'] = df_titanic['Cabin'].map(df_titanic.groupby('Cabin').size())
df_titanic['Embarked'] = df_titanic['Embarked'].map(df_titanic.groupby('Embarked').size())
```

```python
df_titanic
```

```python
simple_train = df_titanic[columns].loc[:500]
simple_test = df_titanic[columns].loc[500:700]
simple_val = df_titanic[columns].loc[700:]
```

```python
max_train = simple_train.max(axis = 0)
min_train = simple_train.min(axis = 0)

simple_train = (simple_train - min_train) / (max_train - min_train)

max_test = simple_test.max(axis = 0)
min_test = simple_test.min(axis = 0)

simple_test = (simple_test - min_test) / (max_test - min_test)

max_val = simple_val.max(axis = 0)
min_val = simple_val.min(axis = 0)

simple_val = (simple_val - min_val) / (max_val - min_val)
```

```python
X_train = simple_train.drop('Survived', axis=1).values
X_test = simple_test.drop('Survived', axis=1).values
X_val = simple_val.drop('Survived', axis=1).values
y_train = simple_train['Survived'].values
y_test = simple_test['Survived'].values
y_val = simple_val['Survived'].values
```

```python
print(X_test)
print(X_train)
print(X_val)
print(y_test)
print(y_train)
print(y_val)
```

```python
best = (0, 0.0)
for par in range(1, 25):
    knn = KNN(par)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_val)
    acc = np.mean(y_pred == y_val)
    print('knn parameter = {} : % = {}'.format(par, acc))
    if acc > best[1]:
        best = (par, acc)

print('best parameter is {}'.format(best[0]))
```

```python
knn = KNN(best[0])
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('процент правильных ответов = {}'.format(np.mean(y_pred == y_test)))
```
